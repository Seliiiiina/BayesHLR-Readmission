```python
import pandas as pd
import numpy as np

data = pd.read_csv("./data/diabetic_data.csv")
data

print("shape:", data.shape)
print("the number of the unique encounter:", data["encounter_id"].nunique())
print("the number of the unique patient:", data["patient_nbr"].nunique())
```

```python
# Section 1: Data Cleaning and Preprocessing
# ============================================================

# Step 1. Check for duplicate encounter records
dup_enc = data.duplicated(subset=["encounter_id"]).sum()
print("Number of duplicated encounter_id records:", dup_enc)

# Step 2. Check for missing records
data = data.replace("?", np.nan).copy()
missing_rate = data.isna().mean().sort_values(ascending=False)
missing_rate[missing_rate > 0]
```

```python
"""
Missingness in weight, max_glu_serum, and A1Cresult is belong to Missing Not At Random (MNAR), as the decision to measure these clinical variables may depend on patients’ underlying health status or disease severity. Missingness in medical_specialty and payer_code belong to Missing At Random (MAR), as it is driven by administrative or documentation processes rather than unobserved patient risk. Missingness in weight and diagnosis codes is belong to structural absence.
"""

data.drop(columns=['weight', 'payer_code'],errors='ignore',inplace=True)

data["max_glu_missing"] = data["max_glu_serum"].isna().astype(int)
data["A1C_missing"] = data["A1Cresult"].isna().astype(int)
data["A1Cresult"] = data["A1Cresult"].fillna("NotMeasured")
data["max_glu_serum"] = data["max_glu_serum"].fillna("NotMeasured")

data["race"] = data["race"].fillna("Unknown")
data["medical_specialty"] = data["medical_specialty"].fillna("Unknown")
```

```python
diag_cols = ["diag_1","diag_2","diag_3"]
for c in diag_cols:
    data[c] = data[c].astype(str).replace("nan", np.nan)

# Step 3. Outcome Refinition
data["readmitted"] = (data["readmitted"] == "<30").astype(int)
print("the rate of the duplicated encounter:", data["readmitted"].mean())
```

```python
def extract_icd_num(code):
    if pd.isna(code):
        return np.nan
    
    code = str(code).strip()
    
    if code[0].isdigit():
        try:
            return float(code[:3])
        except:
            return np.nan
    return np.nan

def icd9_group(num):
    if pd.isna(num):
        return "Other"
    elif 1 <= num <= 139:
        return "Infectious"
    elif 140 <= num <= 239:
        return "Neoplasms"
    elif 240 <= num <= 279:
        return "Endocrine"
    elif 280 <= num <= 289:
        return "Blood"
    elif 290 <= num <= 319:
        return "Mental"
    elif 320 <= num <= 389:
        return "Nervous"
    elif 390 <= num <= 459:
        return "Circulatory"
    elif 460 <= num <= 519:
        return "Respiratory"
    elif 520 <= num <= 579:
        return "Digestive"
    elif 580 <= num <= 629:
        return "Genitourinary"
    elif 630 <= num <= 679:
        return "Pregnancy"
    elif 680 <= num <= 709:
        return "Skin"
    elif 710 <= num <= 739:
        return "Musculoskeletal"
    elif 740 <= num <= 759:
        return "Congenital"
    elif 760 <= num <= 779:
        return "Perinatal"
    elif 780 <= num <= 799:
        return "Symptoms"
    elif 800 <= num <= 999:
        return "Injury"
    else:
        return "Other"
```

```python
# Section 4: Construct design matrix X and outcome vector y
# ============================================================
# Step 1：Define variable groups
continuous_vars = ['age', 'time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']
categorical_vars = ['race', 'gender', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id', 'change', 'diabetesMed', 'diag_group']
drug_vars = ["metformin","repaglinide","nateglinide","chlorpropamide","glimepiride", "acetohexamide","glipizide","glyburide","tolbutamide","pioglitazone", "rosiglitazone","acarbose","miglitol","troglitazone","tolazamide", "examide","citoglipton","insulin","glyburide-metformin","glipizide-metformin", "glimepiride-pioglitazone","metformin-rosiglitazone","metformin-pioglitazone"]
grouping_vars = ['medical_specialty']

# Step 2：Data cleaning and feature engineering
data["age"] = data["age"].astype(str).str.extract(r"(\d+)-(\d+)").astype(float).mean(axis=1)
data["diag1_num"] = data["diag_1"].apply(extract_icd_num)
data["diag_group"] = data["diag1_num"].apply(icd9_group)
data["diag_group"] = data["diag_group"].astype("category")
data.drop(columns=['patient_nbr', 'encounter_id', "diag_1", "diag_2", "diag_3", "diag1_num"],errors='ignore',inplace=True)

# Step 3：Continuous variable transformation
X_cont = data[continuous_vars].copy()
X_cont = (X_cont - X_cont.mean()) / X_cont.std(ddof=0)

# Step 4: Categorical variable transformation
X_cat = pd.get_dummies(data[categorical_vars], drop_first=True)
X_drug = pd.get_dummies(data[drug_vars], drop_first=True)

# Step 5: Combine all features into design matrix X
X = pd.concat([X_cont, X_cat, X_drug], axis=1)
assert not X.isnull().any().any(), "X contains NaN"

# Step 6：Construct design matrix X and outcome vector y
X_cols = X.columns
X = X.values
N, K = X.shape
print("N =", N)
print("K =", K)
y = data["readmitted"].values

# Step 7：Specialty Index
spec_index, spec_levels = pd.factorize(data["medical_specialty"])
J = len(spec_levels)
print("Number of specialties (J):", J)
```
